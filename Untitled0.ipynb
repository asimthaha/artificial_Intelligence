```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer

def preprocess_categorical_columns(data):
    transformed_data = data.copy()  # Create a copy of the data to avoid modifying the original

    categorical_feature_indexes = []  # Store indexes of categorical columns

    for col_index, col in enumerate(transformed_data.T):
        if col.dtype == 'object':
            categorical_feature_indexes.append(col_index)
            labelencoder = LabelEncoder()
            transformed_data[:, col_index] = labelencoder.fit_transform(col)

    if categorical_feature_indexes:
        column_transformer = ColumnTransformer(
            transformers=[('encoder', OneHotEncoder(), categorical_feature_indexes)],
            remainder='passthrough'
        )
        transformed_data = column_transformer.fit_transform(transformed_data)

    return transformed_data

# Assuming 'x' is your input data
x_preprocessed = preprocess_categorical_columns(x)
print(x_preprocessed)
```

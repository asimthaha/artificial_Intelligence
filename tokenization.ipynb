```python
import nltk.data
```

```python
nltk.download('punkt')
```

```python
#loading punk sentence tokenizer using english picklw file
tokenizer = nltk.data.load('tokenizers/punkt/PY3/english.pickle')

text="Google is a multinational technology company that specializes in Internet-related services and products. It was founded by Larry Page and Sergey Brin while they were Ph.D. students at Stanford University in September 1998. The company's mission is to organize the world's information and make it universally accessible and useful."

tokenizer.tokenize(text)
```

```python
#word tokenization

from nltk.tokenize import word_tokenize
text="The script extracts key information about 55-inch televisions from Flipkart's webpage. Specifically, it captures details such as the TV's name, price, and rating from individual product cards on the webpage"
word_tokenize(text)
```

```python
#without . ,
from nltk.tokenize import TreebankWordTokenizer

text="The script extracts key information about 55-inch televisions from Flipkart's webpage. Specifically, it captures details such as the TV's name, price, and rating from individual product cards on the webpage"

tokenizer= TreebankWordTokenizer()
tokenizer.tokenize(text)
```

```python
from nltk.tokenize.regexp import WordPunctTokenizer
#return the tokens from astring of alphabetic character

WordPunctTokenizer

WordPunctTokenizer().tokenize("The script extracts key information about 55-inch televisions from Flipkart's webpage. Specifically, it captures details such as the TV'")
```

```python
#return array of tokens using REGEX

from nltk.tokenize import RegexpTokenizer

tk=RegexpTokenizer('\s+', gaps=True)
gfg="Let's see how its working"
print(tk.tokenize(gfg))
```

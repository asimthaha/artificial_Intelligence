```python
import spacy

nlp=spacy.load('en_core_web_sm')

doc= nlp("the bats saw the cat with best stripes hanging upside down by their feet")

tokens=[]
for token in doc:
  tokens.append(token)
print(tokens)

lemmatized_sentence = " ".join([token.lemma_ for token in doc])
print(lemmatized_sentence)
```

```python
from nltk.tokenize import TreebankWordTokenizer

text="The script extracts key information about 55-inch televisions from Flipkart's webpage. Specifically, it captures details such as the TV's name, price, and rating from individual product cards on the webpage"

tokenizer= TreebankWordTokenizer()
list=tokenizer.tokenize(text)
```

```python
# stemming words from an input of tokenized words

from nltk.stem import PorterStemmer
nltk.download("punkt")

ps=PorterStemmer()
# ex=['words','programming','programer','programs','hallucination','floccinaucinihilipilification']
print("{0:20}{1:20}".format("--Word--","--Stem--"))

for word in list:
  print("{0:20}{1:20}".format(word,ps.stem(word)))
```

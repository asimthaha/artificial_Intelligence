```python
import numpy as np
img = np.zeros((300,300,3),dtype="uint8")
plt.imshow(img)
print(img.shape)
```

```python
#Draw line
line = cv2.line(img,(20,30),(250,250),(48,191,38),8)
line = cv2.line(img,(10,10),(200,200),(0,0,255),3)
line = cv2.line(line,(10,100),(300,100),(255,0,0),5)
plt.imshow(line)
```

```python
#arrowed line
line=cv2.arrowedLine(img,(10,10),(200,200),(0,0,255),6)
plt.imshow(line)
```

```python
#draw rectangle
img=np.zeros((300,300,3),dtype="uint8")
rect1=cv2.rectangle(img,(50,10),(250,250),(155,171,32),6)

plt.imshow(rect1)
```

```python
#draw circle
img=np.zeros((300,300,3),dtype="uint8")
circle=cv2.circle(img,(150,150),100,(219,125,18),6)
circle=cv2.circle(img,(150,150),500,(219,125,200),1)
plt.imshow(circle)
```

```python
#draw elipse
img=np.zeros((300,300,3),dtype="uint8")
e1= cv2.ellipse(img,(200,200),(150,50),0,0,180,255,-1)
e1=cv2.ellipse(img,(200,100),(150,50),30,0,360,(163,20,44),5)
plt.imshow(e1)
```
